3) Create a list of categories and another list of corresponding values. (0.2 p)
- Plot the values as a bar chart with the categories on the x-axis.
- Customize the bar chart with different colors and labels.
- Add a title to the plot and label the x and y axes.
- Display the plot.

```python
import matplotlib.pyplot as plt

categories = ["Category1", "Category2", "Category3", "Category4"]
values = [10, 20, 15, 25]

plt.bar(categories, values, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Customized Bar Chart')
plt.show()
```

6) Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list. (0.15 p)

```python
def sum_even_numbers(numbers):
    even_sum = sum(num for num in numbers if num % 2 == 0)
    return even_sum

numbers = [1, 2, 3, 4, 5, 6]
result = sum_even_numbers(numbers)
print(result)  
```

7. Use a lambda function to filter out all the words from a list of strings that start with the letter 'A'. (0.15 p)

```python
words = ["Apple", "Banana", "Cherry", "Apricot", "Grapes"]

filtered_words = list(filter(lambda word: word.startswith('A'), words))

print(filtered_words) 
```
