```python
import seaborn as sns
import matplotlib.pyplot as plt

# load the tips dataset present by default in seaborn
tips = sns.load_dataset('tips')
#white, dark, whitegrid, darkgrid, ticks
sns.set_style('white')

# make a countplot
sns.countplot(x ='sex', data = tips)
```

```python
sns.set(style='whitegrid')
fmri = sns.load_dataset("fmri")
 
sns.scatterplot(x="timepoint",
                    y="signal",
                    hue="region",
                    style="event",
                    palette='pastel',
                    data=fmri)
```

```python
sns.set(style='whitegrid')
tip = sns.load_dataset('tips')
sns.scatterplot(x='day', y='tip', data=tip ,hue='size', size = "size")
```

```python
sns.scatterplot(x='day', y='tip', data=tip, alpha = 0.1)
```

```python
# import data
data = sns.load_dataset('healthexp')

 
# selecting required rows and columns
data = data.loc[:, ['Year', 'Spending_USD']]
 
# plotting a single line graph
sns.lineplot(x="Year", y="Spending_USD", data=data)
 
# displaying the plot
plt.show()
```

```python
# Import necessary libraries
import seaborn as sns
import numpy as np
import pandas as pd

# Generating dataset of random numbers
np.random.seed(1)
num_var = np.random.randn(1000)
num_var = pd.Series(num_var, name = "Numerical Variable")

# Plot histogram
sns.histplot(data = num_var, kde = True)
```

```python

# data x and y axis for seaborn
x= np.random.randn(200)
y = np.random.randn(200)

sns.kdeplot(x, shade = True , color = "Green")
```

```python
  
# generating 2-D 10x10 matrix of random numbers
# from 1 to 100
data = np.random.randint(low = 1,
                         high = 100,
                         size = (10, 10))
print("The data to be plotted:\n")
print(data)
  
# plotting the heatmap
hm = sns.heatmap(data = data)
  
# displaying the plotted heatmap
plt.show()
```
