# <center><font color='##d0e024'>üßπ 2.Data Cleaning ‚Äî Handling Missing Data</font></center>

---

## <font color='##d0e024'>‚ùì What is Missing Data?</font>

**Missing Data** occurs when no value is stored for a data point in a dataset. It‚Äôs a common issue in real-world datasets that can affect data analysis and machine learning models.

> ‚ö†Ô∏è **‚ÄúMissing data can lead to inaccurate conclusions if not handled properly.‚Äù**

### <font color='##d0e024'>üí° Why Does It Happen?</font>

- üìù **Human Errors:** Incomplete data entry  
- üìä **Data Collection Issues:** Sensor failures, system crashes  
- üïµÔ∏è‚Äç‚ôÇÔ∏è **Privacy Concerns:** Users may intentionally skip answers  
- üìÜ **Time Lags:** Data yet to be recorded

---

## <font color='##d0e024'>üìÇ Types of Missing Data</font>

1. **MCAR ‚Äî Missing Completely at Random**  
   - No relationship between missing data and any variables.  
   - *Example:* A survey form lost due to a system error.

2. **MAR ‚Äî Missing at Random**  
   - Missingness depends on observed data but not on the missing data itself.  
   - *Example:* People from a certain age group are less likely to answer income questions.

3. **MNAR ‚Äî Missing Not at Random**  
   - Missingness depends on unobserved data.  
   - *Example:* Patients not reporting weight due to personal concerns.

# <center><font color='#4C4C9D'>üìä 4.Handling Outliers</font></center>

---

## <font color='#4C4C9D'>üí° What Are Outliers?</font>

**Outliers** are data points that significantly differ from other observations.  
They can occur due to:  
- **Data entry errors**  
- **Measurement variability**  
- **Natural deviations** in the data  

‚ö†Ô∏è **Why Care?**  
Outliers can **skew analysis**, impact **model accuracy**, and lead to **misleading insights**.

---

## <font color='#4C4C9D'>üîç Outlier Detection Methods</font>

### üì¶ **1. Boxplots (IQR Method)**  
- Detect outliers using the **Interquartile Range (IQR)**.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
df = pd.DataFrame({'Score': [45, 48, 50, 52, 47, 200, 49, 53, 51, 46]})

# Boxplot Visualization
sns.boxplot(x=df['Score'])
plt.title('Boxplot for Outlier Detection')
plt.show()

# IQR Calculation
Q1 = df['Score'].quantile(0.25)
Q3 = df['Score'].quantile(0.75)
IQR = Q3 - Q1

# Outlier Boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detecting Outliers
outliers = df[(df['Score'] < lower_bound) | (df['Score'] > upper_bound)]
outliers


````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
df = pd.DataFrame({'Score': [45, 48, 50, 52, 47, 200, 49, 53, 51, 46]})

# Boxplot Visualization
sns.boxplot(x=df['Score'])
plt.title('Boxplot for Outlier Detection')
plt.show()

# IQR Calculation
Q1 = df['Score'].quantile(0.25)
Q3 = df['Score'].quantile(0.75)
IQR = Q3 - Q1

# Outlier Boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detecting Outliers
outliers = df[(df['Score'] < lower_bound) | (df['Score'] > upper_bound)]
outliers
````

# <center><font color='#2A9D8F'>üìä 2.Scatter Plot</font></center>

## üîç **Why Use a Scatter Plot for Outlier Detection?**
A scatter plot helps identify **outliers** by visualizing relationships between two numerical variables.  
Outliers typically appear as points far from the main cluster, making them easy to spot.

---

## üíª **Python Code:**
```python



````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data with Outliers
np.random.seed(42)
x = np.random.normal(50, 10, 100)  # Normally distributed data
y = np.random.normal(50, 10, 100)

# Adding outliers
x = np.append(x, [120, 130, 140])
y = np.append(y, [120, 130, 140])

# Create DataFrame
df = pd.DataFrame({'X': x, 'Y': y})

# Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['X'], y=df['Y'], color='blue', label='Data Points')

# Highlight Outliers (values > 100 in this case)
outliers = df[(df['X'] > 100) | (df['Y'] > 100)]
sns.scatterplot(x=outliers['X'], y=outliers['Y'], color='red', label='Outliers', s=100, marker='X')

plt.title('Scatter Plot for Outlier Detection')
plt.xlabel('Feature X')
plt.ylabel('Feature Y')
plt.legend()
plt.show()

````

# <center><font color='#4C4C9D'>üìä  5.Z-Score Method</font></center>

---

## <font color='#4C4C9D'>üí° What is Z-Score?</font>
The **Z-Score** measures how many **standard deviations** a data point is from the mean.  
- Data points with **|Z-Score| > 3** are typically considered **outliers**.

---

### üßÆ **Z-Score Calculation Example**

```python
from scipy import stats
import pandas as pd
import numpy as np

# Sample DataFrame
df = pd.DataFrame({'Score': [45, 48, 50, 52, 47, 200, 49, 53, 51, 46]})

# Z-Score Calculation
df['Z_Score'] = stats.zscore(df['Score'])

# Outliers: Z-Score > 3 or < -3
df_outliers = df[np.abs(df['Z_Score']) > 3]
df_outliers


````python
from scipy import stats
import pandas as pd
import numpy as np

# Sample DataFrame
df = pd.DataFrame({'Score': [45, 48, 50, 52, 47, 200, 49, 53, 51, 46]})

# Z-Score Calculation
df['Z_Score'] = stats.zscore(df['Score'])

# Outliers: Z-Score > 3 or < -3
df_outliers = df[np.abs(df['Z_Score']) > 3]
df_outliers
````

# <center><font color='#2A9D8F'>üìè 6.Feature Scaling in Machine Learning</font></center>

---

## <font color='#2A9D8F'>üí° Why is Feature Scaling Important?</font>

Machine learning algorithms often rely on **distance-based calculations** (e.g., KNN, SVM, Gradient Descent).  
Unscaled data can lead to **biased models** where features with larger ranges dominate others.

üîë **Key Reasons to Scale Data:**
- ‚öñÔ∏è Ensures **fair weight** across features  
- üöÄ Improves **convergence speed** in gradient-based models  
- üìä Enhances **accuracy** for distance-based algorithms (e.g., KNN, SVM)

---

# <center><font color='#2A9D8F'>üîÑ Types of Feature Scaling</font></center>

| **Scaling Method**   | üìè **Definition**                    | üìä **Best For**                   |
|---------------------|--------------------------------------|----------------------------------|
| **Standardization**  | Centers data (Mean = 0, Std = 1)    | Linear Regression, PCA, SVM     |
| **Normalization**    | Scales to range [0, 1]              | KNN, Neural Networks            |
| **Log Transformation** | Compresses large ranges            | Skewed Distributions            |

## **üìñ Definition & Importance**  
Categorical variables contain discrete values (*e.g., Gender, Country, Product Type*) that machine learning models cannot process directly. **Encoding** transforms these categories into numerical values, making them suitable for algorithms.  

**Why is it important?**  
- üöÄ **Improves model performance**  
- ü§ñ **Enables algorithms to interpret data correctly**  
- üìä **Prevents bias caused by categorical ordering**  

---

# <center><font color='#2A9D8F'>üé® 8.1.2 One-Hot Encoding</font></center>

## üìñ **Definition:**  
One-Hot Encoding transforms **categorical variables** into **binary columns** (0 or 1) for each unique category. It‚Äôs ideal for **nominal data** where no natural ordering exists (*e.g., Country, Color*).

---

## üß† **Why Use One-Hot Encoding?**  
- ‚úÖ **Prevents false ordinal relationships**  
- üìä **Makes data interpretable for ML models**  
- üöÄ **Works well with distance-based algorithms (e.g., KNN, SVM)**  

---

## üêç **Python Example (Pandas)**

```python

# <center><font color='#F4A261'>üéØ 8.1.3 Target Encoding</font></center>

## üìñ **Definition:**  
**Target Encoding** replaces each category with the **mean of the target variable** for that category. It‚Äôs highly effective for **high-cardinality features** (*e.g., cities, user IDs*) and works well with **tree-based models**.

---

## üß† **When to Use Target Encoding?**  
- üìä **High-cardinality features** (many unique values)  
- üå≥ **Tree-based models** (*Random Forest, XGBoost*)  
- ‚öñÔ∏è **When dimensionality needs to be controlled**  

üí° **Avoid using with linear models** without regularization ‚Äî it may introduce **data leakage**.

---

# üî∏ **Wrapper Methods (Use predictive models to score feature subsets)**

**Recursive Feature Elimination (RFE):**  
RFE is a recursive method that removes less significant features based on model performance, aiming to select features that contribute the most to prediction accuracy.

---

## ‚ö° **How It Works:**  
1. Trains a model on all features.  
2. Evaluates feature importance using model coefficients.  
3. Eliminates the least important feature(s).  
4. Repeats steps 1-3 until the desired number of features remains.

---

### üõ†Ô∏è **Example Using RFE with Logistic Regression**

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import pandas as pd

# Sample Data
df = pd.DataFrame({
    'Feature1': [1, 2, 3, 4, 5],
    'Feature2': [5, 4, 3, 2, 1],
    'Feature3': [2, 3, 4, 5, 6],
    'Target': [0, 1, 0, 1, 0]
})

X = df[['Feature1', 'Feature2', 'Feature3']]
y = df['Target']

# Logistic Regression Model
model = LogisticRegression()

# Apply RFE to select the top feature
rfe = RFE(estimator=model, n_features_to_select=1)
fit = rfe.fit(X, y)

# Display Results
print("Num Features Selected:", fit.n_features_)
print("Selected Features:", fit.support_)
print("Feature Ranking:", fit.ranking_)


````python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import pandas as pd

# Sample Data
df = pd.DataFrame({
    'Feature1': [1, 2, 3, 4, 5],
    'Feature2': [5, 4, 3, 2, 1],
    'Feature3': [2, 3, 4, 5, 6],
    'Target': [0, 1, 0, 1, 0]
})

X = df[['Feature1', 'Feature2', 'Feature3']]
y = df['Target']

# Logistic Regression Model
model = LogisticRegression()

# Apply RFE to select the top feature
rfe = RFE(estimator=model, n_features_to_select=1)
fit = rfe.fit(X, y)

# Display Results
print("Num Features Selected:", fit.n_features_)
print("Selected Features:", fit.support_)
print("Feature Ranking:", fit.ranking_)
````

# üî∏ **Embedded Methods (Feature selection during model training)**

**Lasso Regression (L1 Regularization):**  
Lasso adds a penalty equal to the absolute value of the magnitude of coefficients, effectively shrinking less important feature coefficients to zero. This results in a simpler, more interpretable model by automatically selecting features.

---

## ‚ö° **How It Works:**  
- Adds **L1 regularization** term to the loss function.  
- Coefficients of less important features are reduced to zero.  
- Only features with non-zero coefficients are retained.

**Lasso Objective Function:**  
$$
\text{Minimize} \left( \sum_{i=1}^{n}(y_i - \hat{y}_i)^2 + \alpha \sum_{j=1}^{p} |w_j| \right)
$$  
Where:  
- \( \alpha \) controls the strength of regularization.  
- Higher \( \alpha \) ‚Üí more coefficients set to zero.

---

### üõ†Ô∏è **Example Using Lasso Regression for Feature Selection**

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
import pandas as pd

# Sample Data
df = pd.DataFrame({
    'Feature1': [1, 2, 3, 4, 5],
    'Feature2': [5, 4, 3, 2, 1],
    'Feature3': [2, 3, 4, 5, 6],
    'Target': [0, 1, 0, 1, 0]
})

X = df[['Feature1', 'Feature2', 'Feature3']]
y = df['Target']

# Apply Lasso Regression
lasso = Lasso(alpha=0.01)
lasso.fit(X, y)

# Select important features
model = SelectFromModel(lasso, prefit=True)
X_selected = model.transform(X)

# Output selected features
print("Lasso Selected Features:\n", X_selected)

# üìä **11.Log & Power Transformations**

## üìñ **Definition:**  
Log and power transformations are mathematical techniques used to **reshape data distributions**, often to reduce skewness and make data more suitable for machine learning algorithms that assume normality (e.g., linear regression).

---

## üí° **When to Use:**  
- **Skewed Data:** Helps in reducing right or left skewness.  
- **Non-Normal Distributions:** Many ML models perform better with normally distributed data.  
- **Heteroscedasticity:** Stabilizes variance in regression models.  
- **Outliers:** Reduces the impact of extreme values.

---

## üßÆ **Mathematical Explanation:**  

### üî∏ **Log Transformation:**  
Reduces right skewness by compressing large values.  
$$
X' = \log(X + c)
$$  
*(Add a constant \( c \) if \( X \) contains zero or negative values)*

### üî∏ **Square Root Transformation:**  
Reduces moderate right skewness.  
$$
X' = \sqrt{X}
$$

### üî∏ **Box-Cox Transformation:**  
A flexible method that includes log and power transforms.  
$$
X' = \frac{X^\lambda - 1}{\lambda}, \text{ if } \lambda \neq 0  
$$  
$$
X' = \log(X), \text{ if } \lambda = 0
$$

---

## üõ†Ô∏è **Code Example: Log & Box-Cox Transformations**

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Sample Data (Skewed)
data = pd.DataFrame({'Sales': [100, 200, 500, 3000, 150, 6000, 400]})

# Log Transformation
data['Sales_log'] = np.log1p(data['Sales'])

# Box-Cox Transformation (Requires positive data)
data['Sales_boxcox'], fitted_lambda = stats.boxcox(data['Sales'])

# Plot Distributions
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].hist(data['Sales'], bins=20, color='skyblue')
axes[0].set_title('Original Distribution')

axes[1].hist(data['Sales_log'], bins=20, color='lightgreen')
axes[1].set_title('Log Transformed')

axes[2].hist(data['Sales_boxcox'], bins=20, color='salmon')
axes[2].set_title('Box-Cox Transformed')

plt.show()
