```python
import requests
from bs4 import BeautifulSoup
from deep_translator import GoogleTranslator

course_ids = []
for i in range(1, 24):
    course_list_url = f"https://stepik.org:443/api/course-lists?page={i}"
    response = requests.get(course_list_url)
    
    if response.status_code == 200:
        data = response.json()
        if "course-lists" in data:
            for course_list in data["course-lists"]:
                if "courses" in course_list:
                    course_ids.extend(course_list["courses"])
        else:
            print("–ö–ª—é—á 'course-lists' –Ω–µ –Ω–∞–π–¥–µ–Ω –≤ –æ—Ç–≤–µ—Ç–µ.")
    else:
        print("–û—à–∏–±–∫–∞ –ø—Ä–∏ –ø–æ–ª—É—á–µ–Ω–∏–∏ —Å–ø–∏—Å–∫–∞ –∫—É—Ä—Å–æ–≤:", response.status_code, i)
print("–ù–∞–π–¥–µ–Ω–æ –∫—É—Ä—Å–æ–≤:", len(set(course_ids)))
```
