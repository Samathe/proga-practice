```python
# Read the CSV file without a header
df = pd.read_csv("Edx_course_links.csv", header=None)

# Use a list comprehension to create a list of links from the first (and only) column
links_list = [link for link in df[0]]

print(links_list[1:5])
```

```python
links_list = list(dict.fromkeys(links_list))
print(len(links_list))
```

```python
def get_price(driver):
    try:
        price_element = driver.find_element(
            By.CSS_SELECTOR,
            "div.content-center.text-center.text-base.md\\:text-lg.\\!border-t-0"
        )
        if price_element == None:
            return price_element
        elif price_element.text.strip():
            return int(price_element.text.split()[0][1::])
    except Exception:
        ul_elements = driver.find_elements(By.CSS_SELECTOR, "ul.list-none.flex.flex-col.gap-4")
        for ul in ul_elements:
            li_elements = ul.find_elements(By.TAG_NAME, "li")
            if len(li_elements) >= 3:
                li = li_elements[3]
                try:
                    span_element = li.find_element(By.CSS_SELECTOR, "span.mt-0.font-bold.text-lg")
                    return span_element.text.split('\n')[1]
                except Exception:
                    return None
                   

def get_bg_block(driver):  
    try:
         return driver.find_element(
        By.CSS_SELECTOR,
        "div.bg-block.text-primary.py-4.px-2.flex.flex-wrap"
         )
    except Exception:
        pass
    return None
    
def get_inst(driver):
    """
    Finds the first <div> element with class "w-[180px] px-4" and inside it, 
    returns the alt attribute value of the <img> element.
    If the element or alt attribute is not found, returns None.
    """
    try:
        # Use CSS selector with escaped square brackets
        div_element = driver.find_element(By.CSS_SELECTOR, "div.w-\\[180px\\].px-4")
        img_element = div_element.find_element(By.TAG_NAME, "img")
        alt_text = img_element.get_attribute("alt")
        return alt_text
    except Exception as e:
        print(f"Error: {e}")
        return None

    
def get_course_rating(driver):
    """
    Ищет все элементы <div> с классом "col-span-12 md:col-span-7" и возвращает список словарей,
    где для каждого элемента указаны:
      - rating: рейтинг курса,
        - survey_count: сколько людей опросили.
    
    Если строк в элементе меньше трёх, недостающие значения подставляются как None.
    """
    try:
        course_divs = driver.find_elements(By.CSS_SELECTOR, "div.col-span-12.md\\:col-span-7")
        courses = []
        for div in course_divs:
            text = div.text.strip()
            if not text:
                continue
            lines = text.split("\n")
            try: 
                course_name = lines[0] 
                rating = lines[3] 
                survey_count = lines[4]
                courses.append({
                    "rating": rating,
                    "survey_count": survey_count
                })
            except Exception as e:
                courses.append({
                    "rating": None,
                    "survey_count": None
                })
        return courses
    except Exception:
        return None
    
def get_duration(driver, timeout=5):
    try:
        duration_element = WebDriverWait(driver, timeout).until(
            EC.presence_of_element_located(
                (By.CSS_SELECTOR, "div.md\\:text-left.flex-grow.basis-1\\/3.pb-4.md\\:pb-0.grow-0")
            )
        )
        text = duration_element.text.strip()
        if text:
            return text.split('\n')[0]
    except: 
        span_texts = []
        ul_elements = driver.find_elements(By.CSS_SELECTOR, "ul.list-none.flex.flex-col.gap-4")
        for ul in ul_elements:
            li_elements = ul.find_elements(By.TAG_NAME, "li")
            if len(li_elements) >= 3:
                li = li_elements[2]
                try:
                    span_element = li.find_element(By.CSS_SELECTOR, "span.mt-0.font-bold.text-lg")
                    return span_element.text
                except Exception:
                    return None 
        
def get_title(driver):
    """
    Returns the text of the first <h1> element found on the page.
    If no <h1> element is found, returns None.
    """
    title_element = driver.find_element(By.TAG_NAME, "h1")
    title = title_element.text.strip()
    return title if title else None
   

# Настройка ChromeDriver
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
    "AppleWebKit/537.36 (KHTML, like Gecko) "
    "Chrome/90.0.4430.93 Safari/537.36"
)
service = Service(executable_path=r"C:\Users\anuar\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe")
driver = webdriver.Chrome(service=service, options=chrome_options)

# Список для накопления результатов
results = []
k = 0
for link in links_list:
    k += 1
    cleaned_link = link.strip()
    try:
        driver.get(cleaned_link)
    except Exception as e:
        continue

    time.sleep(2)  # ожидание загрузки страницы

    # Получаем цену (при наличии)
    try:
        price = get_price(driver)
    except Exception as e:
        price = None
        print(cleaned_link, 'Problem with price', e)

    Institution = Subject = Level = Associated_skills = None
    try:
        description_element = get_bg_block(driver)
        if description_element:
            text = description_element.text
            description_text = text.split("\n")
            if len(description_text) >= 5:
                Institution = description_text[0].split(":")[1] 
                Subject = description_text[1].split(":")[1]
                Level = description_text[2].split(":")[1]
                Associated_skills = description_text[-1].split(":")[1]
    except Exception as e:
        Institution = Subject = Level = Associated_skills = None
    if Institution == None:
        Institution = get_inst(driver)
    # Получаем название курса, рейтинг и количество отзывов
    course_name = rating = survey_count = None
    try:
        title_rating = get_course_rating(driver)
        if title_rating and isinstance(title_rating, list) and len(title_rating) > 0:
            rating = title_rating[0].get("rating")
            survey_count = title_rating[0].get("survey_count")
    except Exception as e:
        print(cleaned_link, 'rating problem')

    # Получаем длительность курса
    try:
        duration = get_duration(driver)
    except: 
        print(cleaned_link, 'Duration problem')
        duration = None
    try:    
        course_name = get_title(driver)
    except: 
        print(cleaned_link, "Title problem")
    # Формируем запись для текущей ссылки
    
    record = {
        "url": cleaned_link,
        "course_name" : course_name,
        "rating": rating if rating == None else rating.split()[0] ,
        "survey_count": survey_count if survey_count == None else survey_count.split()[0] ,
        "Institution": Institution,
        "Subject": Subject,
        "Level": Level,
        "Associated_skills": Associated_skills if 'Associated skills' in text else None,
        "price": price,
        "duration": duration
    }
    
    # Считаем количество полей (кроме url), где значение равно None
    none_count = sum(1 for key, value in record.items() if key != "url" and value is None)
    if none_count <= 5:
        results.append(record)
    else: 
        print(k , '\n', record )
    if k % 50 == 0:
        df = pd.DataFrame(results)
        df.to_csv("courses_results.csv", index=False)
        
      

# Сохраняем результаты в CSV после обработки всех ссылок
df = pd.DataFrame(results)
df.to_csv("courses_results.csv", index=False)

driver.quit()
df
```
