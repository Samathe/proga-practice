```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select the numerical columns for correlation
numeric_cols = ['rating', 'survey_count', 'price', 'duration_days']
corr = df[numeric_cols].corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Prepare the data
X = df[['duration_days']].values  # Predictor variable (as a 2D array)
y = df['price'].values            # Response variable

# Fit the linear regression model
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Print the model's intercept and coefficient
print("Intercept:", lin_reg.intercept_)
print("Coefficient:", lin_reg.coef_[0])

# Generate a range of duration values for plotting the regression line
X_range = np.linspace(X.min(), X.max(), 200).reshape(-1, 1)
y_pred = lin_reg.predict(X_range)

# Plot the scatter of the data and the linear regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, y, alpha=0.5, color='blue', label='Data Points')
plt.plot(X_range, y_pred, color='red', linewidth=2, label='Linear Regression Line')
plt.title("Linear Regression: Duration (Days) vs Price")
plt.xlabel("Duration (Days)")
plt.ylabel("Price")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import re

# Define a set of stop words to exclude
stop_words = {
    "and", "or", "the", "in", "of", "to", "for", "a", "an", "is", 
    "are", "at", "from", "that", "by", "with", "on", "skills", "youll", "gain", "learning", "systems", "creat"
}

# Aggregate text from the Associated_skills column after cleaning
text_data = ""
for row in df['Associated_skills']:
    # Convert to lowercase and remove punctuation
    cleaned_text = re.sub(r'[^\w\s]', '', row.lower())
    # Split into individual words and filter out stop words
    tokens = [word for word in cleaned_text.split() if word not in stop_words]
    text_data += " " + " ".join(tokens)

# Create a word cloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_data)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Associated Skills")
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the order you want to see the levels
level_order = ["Beginner", "Introductory", "Intermediate", "Advanced"]

# Filter the DataFrame to keep only these known levels
df_levels = df[df["Level"].isin(level_order)].copy()

# Map each level to an index for sorting
level_map = {level: i for i, level in enumerate(level_order)}
df_levels["level_idx"] = df_levels["Level"].map(level_map)

# Group by 'level_idx' and compute average values
agg_df = df_levels.groupby("level_idx").agg({
    "price": "mean",
    "duration_days": "mean",
}).reset_index()

plt.figure(figsize=(10, 6))

# Plot each metric as a separate line
plt.plot(agg_df["level_idx"], agg_df["price"], marker='s', label="Price($)")

# Replace x-ticks with the actual level names
plt.xticks(agg_df["level_idx"], [level_order[idx] for idx in agg_df["level_idx"]])
plt.xlabel("Course Level")
plt.ylabel("Value")
plt.title("Trend Analysis Across Course Levels")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the order you want to see the levels
level_order = ["Beginner", "Introductory", "Intermediate", "Advanced"]

# Filter the DataFrame to keep only these known levels
df_levels = df[df["Level"].isin(level_order)].copy()

# Map each level to an index for sorting
level_map = {level: i for i, level in enumerate(level_order)}
df_levels["level_idx"] = df_levels["Level"].map(level_map)

# Group by 'level_idx' and compute average values
agg_df = df_levels.groupby("level_idx").agg({
    "price": "mean",
    "duration_days": "mean",
}).reset_index()

plt.figure(figsize=(10, 6))

# Plot each metric as a separate line
plt.plot(agg_df["level_idx"], agg_df["duration_days"], marker='^', label="Duration (days)")

# Replace x-ticks with the actual level names
plt.xticks(agg_df["level_idx"], [level_order[idx] for idx in agg_df["level_idx"]])
plt.xlabel("Course Level")
plt.ylabel("Value")
plt.title("Trend Analysis Across Course Levels")
plt.legend()
plt.tight_layout()
plt.show()
```
