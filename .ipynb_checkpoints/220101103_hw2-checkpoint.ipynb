```python
from scipy.stats import binom

alpha = 0.01
n = 42
p0 = 0.2  
p1 = 0.25  

# Find the critical value c
c = binom.ppf(1 - alpha, n, p0) + 1

# Compute Type II error probability at p1 = 0.25:
beta = binom.cdf(c - 1, n, p1)

print ("Part (a): The probability of a Type II error at p1 = 0.25 is:", beta)
```

```python
alpha = 0.01
p0 = 0.2 
p1 = 0.25  

n = 1  

while True:
    # Determine the critical value c for current n.
    c = binom.ppf(1 - alpha, n, p0) + 1
    # Compute the power at p1:
    power = 1 - binom.cdf(c - 1, n, p1)
    if power >= 0.3:
        break
    n += 1

print("Part (b): The required sample size to achieve a power of 0.3 at p1 = 0.25 is:", n)
```

```python
# --------------------------------------------------------------------
# Data and Computation
# --------------------------------------------------------------------
# Given data:
n1 = 42            # sample size for material A
n2 = 42            # sample size for material B
xbar1 = 0.49       # sample mean for material A
xbar2 = 0.36       # sample mean for material B
s1 = 0.19          # sample standard deviation for material A
s2 = 0.16          # sample standard deviation for material B
delta0 = 0.1       # hypothesized difference under H0

# Degrees of freedom
df = n1 + n2 - 2   # 42 + 42 - 2 = 82

# (1) Compute the pooled standard deviation (s_p)
s_p_squared = (((n1 - 1) * s1**2) + ((n2 - 1) * s2**2)) / float(df)
s_p = sqrt(s_p_squared)

# (2) Compute the Standard Error (SE) of the difference in means:
SE = s_p * sqrt(1.0/n1 + 1.0/n2)

# (3) Compute the t-statistic:
#     t = [ (x̄_A - x̄_B) - δ0 ] / SE
t_stat = ((xbar1 - xbar2) - delta0) / SE

# (4) Compute the one-tailed p-value for H_a: μ_A - μ_B > 0.1:
p_value = 1 - t_dist.cdf(t_stat, df)

# (5) Compute the 95% Confidence Interval for (μ_A - μ_B):
alpha = 0.05
t_crit = t_dist.ppf(1 - alpha/2, df)  # two-tailed critical value for 95% CI
diff = xbar1 - xbar2  # difference in sample means = 0.49 - 0.36 = 0.13
ME = t_crit * SE     # Margin of Error
CI_lower = diff - ME
CI_upper = diff + ME

# --------------------------------------------------------------------
# Output the Results:
# --------------------------------------------------------------------
print("Results:")
print("--------")
print("Pooled Standard Deviation (s_p): {:.4f}".format(s_p))
print("Standard Error (SE): {:.4f}".format(SE))
print("t-statistic: {:.4f}".format(t_stat))
print("Degrees of Freedom: {}".format(df))
print("One-tailed p-value: {:.4f}".format(p_value))
print("95% Confidence Interval for (μ_A - μ_B): ({:.4f}, {:.4f})".format(CI_lower, CI_upper))
```

```python

from __future__ import print_function  # for Python 2/3 compatibility
from math import sqrt
from scipy.stats import t as t_dist

# Given Data:
n1 = 9            # Group 1 (-8°C)
n2 = 9            # Group 2 (15°C)
xbar1 = 62.01     # Sample mean at -8°C
s1 = 3.14         # Sample standard deviation at -8°C
xbar2 = 67.38     # Sample mean at 15°C
s2 = 4.92         # Sample standard deviation at 15°C

# Degrees of freedom:
df = n1 + n2 - 2  # = 9 + 9 - 2 = 16

# (1) Compute the pooled standard deviation, s_p:
#    s_p = sqrt( [ (n1-1)*s1^2 + (n2-1)*s2^2 ] / (n1+n2-2) )
s_p_squared = (((n1 - 1) * s1**2) + ((n2 - 1) * s2**2)) / float(df)
s_p = sqrt(s_p_squared)

# (2) Compute the standard error (SE) of the difference in means:
#    SE = s_p * sqrt(1/n1 + 1/n2)
SE = s_p * sqrt(1.0/n1 + 1.0/n2)

# (3) Compute the difference in sample means:
diff = xbar1 - xbar2   # 62.01 - 67.38 = -5.37

# (4) Compute the t-statistic (under H0: difference = 0):
#    t = ( (X̄1 - X̄2) - 0 ) / SE
t_stat = diff / SE

# (5) Compute the two-sided p-value for the t-test:
# For a two-sided test: p-value = 2 * P(T <= t_stat) if t_stat is negative, else 2 * P(T >= t_stat)
if t_stat < 0:
    p_value = 2 * t_dist.cdf(t_stat, df)
else:
    p_value = 2 * (1 - t_dist.cdf(t_stat, df))

# (6) Construct a 90% Confidence Interval (CI) for the difference in means:
#    CI = (diff) ± t_(1-α/2, df) * SE, with α = 0.10.
alpha = 0.10
t_crit = t_dist.ppf(1 - alpha/2, df)
ME = t_crit * SE  # Margin of Error
CI_lower = diff - ME
CI_upper = diff + ME

# --------------------------------------------------------------------
# Output the results:
# --------------------------------------------------------------------
print("Concrete Strength Analysis (Effect of Temperature)")
print("----------------------------------------------------")
print("Group 1 (-8°C): n =", n1, ", mean =", xbar1, ", std =", s1)
print("Group 2 (15°C): n =", n2, ", mean =", xbar2, ", std =", s2)
print("")
print("Pooled Standard Deviation (s_p): {:.4f}".format(s_p))
print("Standard Error (SE): {:.4f}".format(SE))
print("Difference in Means (X̄1 - X̄2): {:.4f}".format(diff))
print("t-statistic: {:.4f}".format(t_stat))
print("Degrees of Freedom: {}".format(df))
print("Two-sided p-value (α = 0.10): {:.4f}".format(p_value))
print("")
print("90% Confidence Interval for (μ1 - μ2): ({:.4f}, {:.4f})".format(CI_lower, CI_upper))
```

```python
#!/usr/bin/env python
from __future__ import print_function  # for Python 2/3 compatibility
from math import sqrt
from scipy.stats import norm

x1 = 10   # successes for tears >25 mm
n1 = 18
x2 = 22   # successes for tears <25 mm
n2 = 30

# Sample proportions:
p1 = x1 / float(n1)
p2 = x2 / float(n2)
diff = p1 - p2  # difference in proportions

# -------------------------------
# (a) Hypothesis Test (Two-Sided)
# -------------------------------
# Under H0, the success rates are equal so we pool the data:
p_pool = (x1 + x2) / float(n1 + n2)

# Pooled Standard Error:
SE_pooled = sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

# z-statistic for the two-proportion test:
z_stat = diff / SE_pooled

# Two-tailed p-value:
if z_stat < 0:
    p_value = 2 * norm.cdf(z_stat)
else:
    p_value = 2 * (1 - norm.cdf(z_stat))

# -------------------------------
# (b) 90% Confidence Interval for p1 - p2
# -------------------------------
# For the confidence interval we use the unpooled standard error:
SE_unpooled = sqrt(p1 * (1 - p1) / n1 + p2 * (1 - p2) / n2)
# For a 90% CI, the critical z-value for α/2 = 0.05 is:
z_crit = norm.ppf(1 - 0.05)  # approximately 1.645
ME = z_crit * SE_unpooled  # margin of error

CI_lower = diff - ME
CI_upper = diff + ME

# ------------------------------------------------------------
# Output the results:
# ------------------------------------------------------------
print("Arthroscopic Meniscal Repair Analysis:")
print("----------------------------------------")
print("Group 1 (tears >25 mm): successes = {}, n = {}, p1 = {:.4f}".format(x1, n1, p1))
print("Group 2 (tears <25 mm): successes = {}, n = {}, p2 = {:.4f}".format(x2, n2, p2))
print("")
print("Pooled Proportion: {:.4f}".format(p_pool))
print("Pooled Standard Error: {:.4f}".format(SE_pooled))
print("z-statistic: {:.4f}".format(z_stat))
print("Two-tailed p-value (α = 0.10): {:.4f}".format(p_value))
print("")
print("90% Confidence Interval for (p1 - p2): ({:.4f}, {:.4f})".format(CI_lower, CI_upper))
```

```python
import numpy as np
from scipy.stats import mannwhitneyu, norm

# ------------------------------------------------------------
# Data:
# SG (spheroidal graphite) measurements (multiplied by 10,000)
SG = np.array([105, 77, 52, 27, 22, 17, 12, 14, 65])
# CG (compacted graphite) measurements (multiplied by 10,000)
CG = np.array([90, 50, 30, 20, 14, 10, 60, 24, 76])

alpha = 0.05  # significance level for the test

# ------------------------------------------------------------
# (a) Mann–Whitney Rank-Sum (Wilcoxon) Test:
#    Null Hypothesis (H0): The distributions for SG and CG are identical
#    Alternative (Ha): They differ (two-sided test)
# ------------------------------------------------------------
# Use SciPy's mannwhitneyu. Note: newer versions allow a 'two-sided' alternative.
result = mannwhitneyu(SG, CG, alternative='two-sided', method='exact')
U_stat = result.statistic
p_value = result.pvalue

# ------------------------------------------------------------
# (b) 95% Confidence Interval for the Median Difference (Hodges–Lehmann estimator)
#    We compute all pairwise differences (SG[i] - CG[j]) and then take the median,
#    and use the 2.5th and 97.5th percentiles as an approximate CI.
# ------------------------------------------------------------

diff_matrix = np.subtract.outer(SG, CG)  # shape (len(SG), len(CG))
all_diffs = diff_matrix.flatten()

# Hodges–Lehmann estimator (median of all differences)
HL_est = np.median(all_diffs)

# Approximate 95% CI (using the empirical percentiles)
CI_lower = np.percentile(all_diffs, 2.5)
CI_upper = np.percentile(all_diffs, 97.5)

# ------------------------------------------------------------
# Output the results:
# ------------------------------------------------------------
print("Mann–Whitney Rank-Sum Test (Wilcoxon) for SG vs. CG")
print("-----------------------------------------------------")
print("U statistic: {:.4f}".format(U_stat))
print("Two-sided p-value: {:.4f}".format(p_value))
print("")
print("Hodges–Lehmann estimator (median difference SG - CG): {:.4f}".format(HL_est))
print("Approximate 95% CI for median difference: ({:.4f}, {:.4f})".format(CI_lower, CI_upper))
print("")
print("Interpretation:")
if p_value < alpha:
    print("  There is evidence (p < {:.2f}) that the total amplitude strain properties differ between SG and CG cast iron.".format(alpha))
else:
    print("  There is no evidence (p >= {:.2f}) that the total amplitude strain properties differ between SG and CG cast iron.".format(alpha))
```
