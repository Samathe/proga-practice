## Clustering results

```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder

# Load the data
df = pd.read_csv('Categorical.csv')
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# Encode the 'continent' column
le = LabelEncoder()
df['continent_encoded'] = le.fit_transform(df['continent'])

# Define the feature matrix for clustering
x = df[['continent_encoded']]

# -------------------------------
# Part 1: Initial Clustering with 4 Clusters
# -------------------------------

# Apply KMeans with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster'] = kmeans.fit_predict(x)

# Print cluster centers and distribution
print("Cluster centers:")
print(kmeans.cluster_centers_)

print("\nCluster distribution:")
print(df['cluster'].value_counts())
```

## Plot the data

```python
# Plot the clustering results using a scatter plot (using DataFrame index as x-axis)
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['continent_encoded'], c=df['cluster'], cmap='viridis', s=100)
plt.xlabel('Country Index')
plt.ylabel('Continent (Encoded)')
plt.title('KMeans Clustering with 4 Clusters')
plt.show()

# -------------------------------
# Part 2: The Elbow Method for Varying Number of Clusters
# -------------------------------

```

```python
# Create an empty list to store the Within-cluster Sum of Squares (WCSS)
wcss = []

# Loop over a range of cluster numbers (1 through 6)
for i in range(1, 7):
    kmeans_iter = KMeans(n_clusters=i, random_state=42)
    kmeans_iter.fit(x)
    wcss.append(kmeans_iter.inertia_)

# Plot the number of clusters vs WCSS
number_clusters = range(1, 7)
plt.figure(figsize=(8, 5))
plt.plot(number_clusters, wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
plt.show()
```
