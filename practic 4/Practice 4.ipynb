```python
import seaborn as sns
import matplotlib.pyplot as plt

data0 = data[data['yr'] == 0]
data1 = data[data['yr'] == 1]

fig = plt.figure()

sns.relplot(x='mnth', y='temp',style="yr", data= data0, kind='line', height=6, markers = True)


sns.lineplot(x = 'mnth', y = 'temp' ,data = data0, markers = True).set(title='Different Temperatures within Months By Year', xlabel='Month', ylabel='Temperature')
sns.lineplot(x = 'mnth', y = 'temp' ,data = data1, markers = True)

title_props = {'fontsize': 12, 'fontweight': 'bold'}

# Create the title and set its position
fig.suptitle('Different Temperatures within Months By Year', **title_props, x=0.55, y=1.0)
# plt.show()
```
