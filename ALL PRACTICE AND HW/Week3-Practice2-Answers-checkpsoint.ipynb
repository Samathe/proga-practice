#### Task 1: Creating a Scatter Plot (0.1 point)

Create two lists, x and y, with random data points.

1. Plot x against y as a scatter plot.
2. Customize the plot with a green color, dashed line and star marker style.
3. Add a title to the plot and label the x and y axes.
4. Display the plot.

```python
import matplotlib.pyplot as plt
import random

# Generate random data
x = [random.randint(1, 10) for _ in range(10)]
y = [random.randint(1, 10) for _ in range(10)]

# Create a scatter plot
plt.scatter(x, y, color='green', marker='*', linestyle = 'dashed')

# Add title and labels
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display the plot
plt.show()

```

#### Task 2: Creating a Bar Chart (0.1 point)

Create a list of categories and another list of corresponding values.

1. Plot the values as a bar chart with the categories on the x-axis.
2. Customize the bar chart with different colors and labels.
3. Add a title to the plot and label the x and y axes.

```python
# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 30, 20, 35]

# Create a bar chart
plt.bar(categories, values, color=['blue', 'green', 'red', 'purple'])

# Add title and labels
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

# Display the plot
plt.show()
```

```python
import numpy as np
# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the points using matplotlib
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
plt.show()
```

#### Task 4: Creating a Histogram (0.1 point)

Generate a list of random data points (e.g., using numpy).

1. Create a histogram to visualize the distribution of the data.
2. Customize the histogram with a different number of bins, color, and alpha value.
3. Add a title to the plot and label the x and y axes.
4. Display the plot.

```python
import numpy as np

# Generate random data
data = np.random.normal(0, 1, 1000)  # 1000 data points with mean 0 and standard deviation 1

# Create a histogram
plt.hist(data, bins=20, color='skyblue', alpha=0.7)

# Add title and labels
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

#### Task 5: Creating a Pie Chart (0.1 point)

Create a list of labels and another list of corresponding values.

1. Plot the values as a pie chart with labels.
2. Customize the pie chart with different colors and explode a slice.
3. Add a title to the plot.
4. Display the plot.

```python
# Sample data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 30, 25, 30]
explode = (0.1, 0, 0, 0)  # Explode the first slice

# Create a pie chart
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', colors=['red', 'green', 'blue', 'purple'])

# Add title
plt.title("Pie Chart")

# Display the plot
plt.show()
```

#### Task 6: Load Data and Display Basic Information (0.1 point)

1. Load the Seaborn built-in dataset 'tips' and assign it to a variable.
2. Display the first 5 rows of the dataset and provide a brief description of the dataset.
3. Check for missing values in the dataset

```python
#Load Data
import seaborn as sns

tips = sns.load_dataset('tips')

#Display Basic Information
print(tips.head())
print(tips.info())

#Check for missing values in the dataset
print(tips.isnull().sum())
```

#### Task 7: Creating a Bar Plot (0.1 point)

Create a bar plot to show the total bill amounts

```python
plt.figure(figsize=(5, 4))
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill Amounts by Day")
plt.show()
```

#### Task 8: Data Exploration (0.1 point)

Create a pair plot to visualize relationships between numerical columns in the dataset.



```python
sns.pairplot(data=tips)
plt.title('Pair Plot of Numerical Columns')
plt.show()
```

#### Task 9: Creating a Box Plot (0.1 point)

Create a box plot to visualize the distribution of 'total_bill' by 'day.'

```python
plt.figure(figsize=(5, 5))
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Total Bill Distribution by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill Amount')
plt.show()
```

#### Task 10: Creating a Scatter Chart (0.1 point)

 Create a scatter plot of 'total_bill' vs 'tip' and observe any patterns.

```python
plt.figure(figsize=(5, 5))
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill Amount')
plt.ylabel('Tip Amount')
plt.show()
```

### SciPy

#### Task 11: Calculate Descriptive Statistics (0.4 point)

Given a dataset of exam scores, calculate the mean, mode, median, standard deviation scores using scipy.stats.

```python
import scipy
import scipy.stats as stats

# Sample data (exam scores)
scores = [85, 92, 78, 90, 88, 76, 95, 89, 92, 87]

# Calculate mean, median, mode, standard deviation
mean_score = scipy.mean(scores)
mode_score = stats.mode(scores)
median_score = scipy.median(scores)
std_dev = scipy.std(scores)

print("Mean:", mean_score)
print("Mode:", mode_score)
print("Median:", median_score)
print("Standard Deviation:", std_dev)
```
