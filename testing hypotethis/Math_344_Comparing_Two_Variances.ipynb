```python
import numpy as np
import scipy.stats as stats

# Data for Nonsmokers and Smokers
nonsmokers = np.array([41.48, 41.71, 41.98, 41.68, 41.18])
smokers = np.array([40.42, 40.68, 40.51, 40.73, 40.91])

# Levene’s Test (Robust to non-normality)
levene_stat, levene_p = stats.levene(nonsmokers, smokers)
print(f"Levene’s Test: W = {levene_stat:.4f}, p-value = {levene_p:.4f}")


# Interpretation
alpha = 0.05
if levene_p > alpha:
    print("Levene’s Test: Fail to reject H0 (variances are equal).")
else:
    print("Levene’s Test: Reject H0 (variances are different).")
```

# $\chi$-squared distribution

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Define degrees of freedom to visualize
df_values = [ 2, 5, 10, 20]
x = np.linspace(0, 30, 1000)  # Range for plotting

# Plot Chi-Square distributions
plt.figure(figsize=(8, 5))
for df in df_values:
    plt.plot(x, stats.chi2.pdf(x, df), label=f'df={df}')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Chi-Square Distributions for Different Degrees of Freedom')
plt.legend()
plt.grid(True)
plt.show()
```

# F distribution

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
# Define degrees of freedom
df1, df2 = 10, 20  # Example values
x = np.linspace(0, 5, 1000)  # Range of F-values

# Compute F-distribution PDF
f_pdf = stats.f.pdf(x, df1, df2)

# Plot F-distribution
plt.figure(figsize=(8, 5))
plt.plot(x, f_pdf, label=f'F({df1}, {df2})', color='b')
plt.fill_between(x, f_pdf, alpha=0.2, color='blue')  # Shaded area
plt.axvline(x=stats.f.ppf(0.95, df1, df2), color='r', linestyle='--', label="95% Critical Value")
plt.xlabel('F-value')
plt.ylabel('Density')
plt.title(f'F-Distribution with df1={df1}, df2={df2}')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
# Define degrees of freedom
df1, df2 = 10, 20  # Degrees of freedom for two chi-square distributions
size = 10000  # Sample size

# Generate Chi-Square distributed data
chi_sq1 = np.random.chisquare(df1, size)
chi_sq2 = np.random.chisquare(df2, size)

# Compute F-ratio
F_values = (chi_sq1 / df1) / (chi_sq2 / df2)

# Plot histogram of F-values
plt.figure(figsize=(8, 5))
plt.hist(F_values, bins=100, density=True, alpha=0.6, color='b', label="Empirical F-Distribution")

# Overlay theoretical F-distribution
x = np.linspace(0, 5, 1000)
f_pdf = stats.f.pdf(x, df1, df2)
plt.plot(x, f_pdf, 'r-', label=f'Theoretical F({df1}, {df2})')

plt.xlabel('F-value')
plt.ylabel('Density')
plt.title('F-Distribution as the Ratio of Two Chi-Square Distributions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as stats


# Data for Nonsmokers and Smokers
nonsmokers = np.array([41.48, 41.71, 41.98, 41.68, 41.18])
smokers = np.array([40.42, 40.68, 40.51, 40.73, 40.91])
# Step 1: Compute sample variances
s1_sq = np.var(nonsmokers, ddof=1)  # Sample variance of first group
s2_sq = np.var(smokers, ddof=1)  # Sample variance of second group

# Step 2: Compute F-statistic
F_stat = s1_sq / s2_sq

# Step 3: Compute degrees of freedom
n1, n2 = len(nonsmokers), len(smokers)
df1, df2 = n1 - 1, n2 - 1

# Step 4: Compute p-value (two-tailed test)
p_value = 2 * min(1 - stats.f.cdf(F_stat, df1, df2), stats.f.cdf(F_stat, df1, df2))

# Step 5: Compute the 95% Confidence Interval for σ1² / σ2²
alpha = 0.05
F_crit_low = stats.f.ppf(alpha / 2, df1, df2)  # Lower critical value
F_crit_high = stats.f.ppf(1 - alpha / 2, df1, df2)  # Upper critical value

ci_lower = F_stat / F_crit_high
ci_upper = F_stat / F_crit_low

# Print results
print(f"F-statistic: {F_stat:.4f}")
print(f"p-value: {p_value:.4f}")
print(f"95% Confidence Interval for σ1² / σ2²: ({ci_lower:.4f}, {ci_upper:.4f})")

# Conclusion based on significance level
if p_value < alpha:
    print("Reject H0: The variances are significantly different.")
else:
    print("Fail to reject H0: No significant difference in variances.")
```

```python
import numpy as np
import scipy.stats as stats


# Data for Nonsmokers and Smokers
nonsmokers = np.array([50.2, 49.8, 50.5, 50.1, 49.7, 50.3])
smokers = np.array([49.5, 50.0, 49.8, 49.7, 50.2, 49.9])
# Step 1: Compute sample variances
s1_sq = np.var(nonsmokers, ddof=1)  # Sample variance of first group
s2_sq = np.var(smokers, ddof=1)  # Sample variance of second group

# Step 2: Compute F-statistic
F_stat = s1_sq / s2_sq

# Step 3: Compute degrees of freedom
n1, n2 = len(nonsmokers), len(smokers)
df1, df2 = n1 - 1, n2 - 1

# Step 4: Compute p-value (two-tailed test)
p_value = 2 * min(1 - stats.f.cdf(F_stat, df1, df2), stats.f.cdf(F_stat, df1, df2))

# Step 5: Compute the 95% Confidence Interval for σ1² / σ2²
alpha = 0.05
F_crit_low = stats.f.ppf(alpha / 2, df1, df2)  # Lower critical value
F_crit_high = stats.f.ppf(1 - alpha / 2, df1, df2)  # Upper critical value

ci_lower = F_stat / F_crit_high
ci_upper = F_stat / F_crit_low

# Print results
print(f"F-statistic: {F_stat:.4f}")
print(f"p-value: {p_value:.4f}")
print(f"95% Confidence Interval for σ1² / σ2²: ({ci_lower:.4f}, {ci_upper:.4f})")

# Conclusion based on significance level
if p_value < alpha:
    print("Reject H0: The variances are significantly different.")
else:
    print("Fail to reject H0: No significant difference in variances.")
```
