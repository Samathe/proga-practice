```python
import numpy as np
from scipy import stats

# Simulate two independent samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=20)  # Sample from population 1
sample2 = np.random.normal(loc=55, scale=15, size=20)  # Sample from population 2

# Sample sizes
n1, n2 = len(sample1), len(sample2)

# Sample means and standard deviations
mean1, mean2 = np.mean(sample1), np.mean(sample2)
std1, std2 = np.std(sample1, ddof=1), np.std(sample2, ddof=1)

# 1. Pooled variance (assuming equal variances)
sp2 = ((n1 - 1) * std1**2 + (n2 - 1) * std2**2) / (n1 + n2 - 2)

# 2. Standard error assuming equal variances
se_pooled = np.sqrt(sp2 * (1/n1 + 1/n2))

# 3. Welchs variance (assuming unequal variances)
se_welch = np.sqrt((std1**2 / n1) + (std2**2 / n2))


# Print results
print(f"Pooled Variance: {sp2:.4f}")
print(f"Standard Error (Pooled Variance): {se_pooled:.6f}")
print(f"Standard Error (Welch's Variance): {se_welch:.6f}")
```

Formula for the Confidence Interval

Since the population variances are unknown, we use the pooled standard deviation and the two-sample t-interval:


$\text{SE} = \sqrt{S_p^2 \left( \frac{1}{n_1} + \frac{1}{n_2} \right)}$


where the pooled variance  $S_p^2$  is:


$S_p^2 = \frac{(n_1 - 1)S_1^2 + (n_2 - 1)S_2^2}{n_1 + n_2 - 2}$


The confidence interval is then given by:


$(X_1 - X_2) \pm t_{\alpha/2, df} \times \text{SE}$


where  $t_{\alpha/2, df}$  is the critical value from the t-distribution with  $df = n_1 + n_2 - 2 .$

```python
import scipy.stats as stats
import numpy as np

# Given data
n1, n2 = 8, 8
X1, S1 = 92.255, 2.39
X2, S2 = 92.733, 2.98

# Compute pooled variance
Sp2 = ((n1 - 1) * S1**2 + (n2 - 1) * S2**2) / (n1 + n2 - 2)
Sp = np.sqrt(Sp2)

# Compute standard error
SE = Sp * np.sqrt(1/n1 + 1/n2)

# Degrees of freedom
df = n1 + n2 - 2

# Critical t-value for 95% confidence level
t_crit = stats.t.ppf(0.975, df)

# Compute margin of error
ME = t_crit * SE

# Compute confidence interval
CI_lower = (X1 - X2) - ME
CI_upper = (X1 - X2) + ME

CI_lower, CI_upper
```

If we assume equal variances, we use the pooled variance:


$S_p^2 = \frac{(n_1 - 1)S_1^2 + (n_2 - 1)S_2^2}{n_1 + n_2 - 2}$


The standard error (SE) is:


$\text{SE} = S_p \times \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}$


The confidence interval formula is:


$(X_1 - X_2) \pm t_{\alpha/2, df} \times \text{SE}$


where  $t_{\alpha/2, df}$  is the critical value from the t-distribution with degrees of freedom:


$df = n_1 + n_2 - 2$

```python
# Given data
n1, n2 = 35, 40
X1, S1 = 24.2, np.sqrt(10.0)
X2, S2 = 23.9, np.sqrt(14.3)

# Compute pooled variance
Sp2 = ((n1 - 1) * S1**2 + (n2 - 1) * S2**2) / (n1 + n2 - 2)
Sp = np.sqrt(Sp2)

# Compute standard error
SE = Sp * np.sqrt(1/n1 + 1/n2)

# Degrees of freedom
df = n1 + n2 - 2

# Critical t-value for 90% confidence level
t_crit = stats.t.ppf(0.95, df)  # One-tailed 5% on each side

# Compute margin of error
ME = t_crit * SE

# Compute confidence interval
CI_lower = (X1 - X2) - ME
CI_upper = (X1 - X2) + ME

CI_lower, CI_upper
```

## Example:
A random sample of $n_1 = 32$ specimens of cold-rolled steel give average strength
$X_1 = 29.80$ ksi, and sample standard deviation of $S_1 = 4.00$ ksi. A random sample of
$n_2 = 35$ specimens of two-sided galvanized steel give average strength $X_2 = 34.70$
ksi, and $S_2 = 6.74$ ksi. Construct a $99\%$ CI for $\mu_1-\mu_2$, and state any assumptions
that are needed for its validity

If we assume equal variances, we use the pooled variance:


$S_p^2 = \frac{(n_1 - 1)S_1^2 + (n_2 - 1)S_2^2}{n_1 + n_2 - 2}$


The standard error (SE) is:


$\text{SE} = S_p \times \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}$


The confidence interval formula is:


$(X_1 - X_2) \pm t_{\alpha/2, df} \times \text{SE}$


where  $t_{\alpha/2, df}$  is the critical value from the t-distribution with degrees of freedom:


$df = n_1 + n_2 - 2$


```python
# Given data
n1, n2 = 32, 35
X1, S1 = 29.80, 4.00
X2, S2 = 34.70, 6.74

# Compute pooled variance
Sp2 = ((n1 - 1) * S1**2 + (n2 - 1) * S2**2) / (n1 + n2 - 2)
Sp = np.sqrt(Sp2)

# Compute standard error
SE = Sp * np.sqrt(1/n1 + 1/n2)

# Degrees of freedom
df = n1 + n2 - 2

# Critical t-value for 99% confidence level
t_crit = stats.t.ppf(0.995, df)  # One-tailed 0.5% on each side

# Compute margin of error
ME = t_crit * SE

# Compute confidence interval
CI_lower = (X1 - X2) - ME
CI_upper = (X1 - X2) + ME

CI_lower, CI_upper
```

## Example
A certain type of tractor is being assembled at two locations, $L_1$ and $L_2$. An investigation into the proportion of tractors requiring extensive adjustments after assembly
finds that in random samples of 200 tractors from $L_1$ and 400 from $L_2$, the number
requiring extensive adjustments were 16 and 14, respectively. Construct a $99\%$ CI
for $p_1 - p_2$, the difference of the two proportions.

Step 1: Define Given Data

Location $L_1:$
$n_1 = 200, \quad x_1 = 16$


$\hat{p}_1 = \frac{x_1}{n_1} = \frac{16}{200} = 0.08$

Location $L_2$: $n_2 = 400, \quad x_2 = 14$


$\hat{p}_2 = \frac{x_2}{n_2} = \frac{14}{400} = 0.035$

Step 2: Compute the Standard Error (SE)

For two independent proportions, the standard error is given by:

$\text{SE} = \sqrt{\frac{\hat{p}_1 (1 - \hat{p}_1)}{n_1} + \frac{\hat{p}_2 (1 - \hat{p}_2)}{n_2}}$

Step 3: Compute the Confidence Interval

The confidence interval formula is:


$(\hat{p}_1 - \hat{p}2) \pm z{\alpha/2} \times \text{SE}$


where:
	$z_{\alpha/2}$  is the critical z-value for a 99% confidence level, which is 2.576.

```python
# Given data
n1, x1 = 200, 16
n2, x2 = 400, 14

# Compute sample proportions
p1_hat = x1 / n1
p2_hat = x2 / n2

# Compute standard error
SE = np.sqrt((p1_hat * (1 - p1_hat) / n1) + (p2_hat * (1 - p2_hat) / n2))

# Critical z-value for 99% confidence level
z_crit = stats.norm.ppf(0.995)  # Two-tailed

# Compute margin of error
ME = z_crit * SE

# Compute confidence interval
CI_lower = (p1_hat - p2_hat) - ME
CI_upper = (p1_hat - p2_hat) + ME

CI_lower, CI_upper
```

```python
import numpy as np
import scipy.stats as stats
import pandas as pd

# Given summary statistics
n1, X1, S1 = 32, 29.8, 4.0
n2, X2, S2 = 35, 34.7, 6.74

# Compute Welch's t-test
t_stat, p_value = stats.ttest_ind_from_stats(mean1=X1, std1=S1, nobs1=n1,
                                             mean2=X2, std2=S2, nobs2=n2,
                                             equal_var=False)

# Compute degrees of freedom for Welch's test
df = ((S1**2/n1 + S2**2/n2)**2) / (((S1**2/n1)**2 / (n1-1)) + ((S2**2/n2)**2 / (n2-1)))

# Compute 99% confidence interval
alpha = 0.01
t_crit = stats.t.ppf(1 - alpha/2, df)  # Critical t-value
SE = np.sqrt(S1**2/n1 + S2**2/n2)  # Standard error
CI_lower = (X1 - X2) - t_crit * SE
CI_upper = (X1 - X2) + t_crit * SE

# Print results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {df:.2f}")
print(f"99% Confidence Interval: ({CI_lower:.4f}, {CI_upper:.4f})")
```

```python
import numpy as np
import scipy.stats as stats
import pandas as pd

# URL of the dataset
url = "https://media.pearsoncmg.com/cmg/pmmg_mml_shared/mathstatsresources/Akritas/SteelStrengthData.txt"

# Read the dataset
df = pd.read_csv(url, delim_whitespace=True)

# Split into two groups based on 'Sample' column
group1 = df[df['Sample'] == 1]['Value']  # ColdRolled steel
group2 = df[df['Sample'] == 2]['Value']  # Galvanized steel

# Perform Welch's t-test (does not assume equal variances)
t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=False)
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")

# Compute standard errors
n1, n2 = len(group1), len(group2)
S1, S2 = np.std(group1, ddof=1), np.std(group2, ddof=1)
X1, X2 = np.mean(group1), np.mean(group2)

SE = np.sqrt(S1**2 / n1 + S2**2 / n2)

# Compute degrees of freedom (Welch-Satterthwaite equation)
df_welch = ((S1**2 / n1 + S2**2 / n2) ** 2) / (((S1**2 / n1) ** 2) / (n1 - 1) + ((S2**2 / n2) ** 2) / (n2 - 1))

# Critical t-value for 99% CI
alpha = 0.01
t_crit = stats.t.ppf(1 - alpha / 2, df_welch)

# Compute confidence interval
CI_lower = (X1 - X2) - t_crit * SE
CI_upper = (X1 - X2) + t_crit * SE

print(f"99% Confidence Interval for mean difference: ({CI_lower:.4f}, {CI_upper:.4f})")
```

```python
import numpy as np
import scipy.stats as stats
import pandas as pd

# Given data
n1, X1, S1 = 35, 24.2, np.sqrt(10)
n2, X2, S2 = 40, 23.9, np.sqrt(14.3)

# Standard error
SE = np.sqrt(S1**2 / n1 + S2**2 / n2)

# Welch-Satterthwaite degrees of freedom
df_welch = ((S1**2 / n1 + S2**2 / n2) ** 2) / (((S1**2 / n1) ** 2) / (n1 - 1) + ((S2**2 / n2) ** 2) / (n2 - 1))

# Compute test statistic (t-value)
t_stat = (X1 - X2) / SE

# Compute one-tailed p-value for H_A: 1 > 2
p_value = 1 - stats.t.cdf(t_stat, df_welch)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Compute 90% Confidence Interval for 1 - 2
alpha = 0.1
t_crit = stats.t.ppf(1 - alpha / 2, df_welch)

CI_lower = (X1 - X2) - t_crit * SE
CI_upper = (X1 - X2) + t_crit * SE

print(f"90% Confidence Interval for mean difference: ({CI_lower:.4f}, {CI_upper:.4f})")
# expected 90% CI (1.052603, 1.652032)
```

## Example 9.2-8
Consider the manufacturing of tractors using two different assembly lines, as
described in Example 9.2-4. Let $p_1$ denote the proportion of tractors coming out
of assembly line $L_1$ that require adjustments, and let $p_2$ be the corresponding
proportion for assembly line $L_2$.

(a) Test $H_0 : p_1 = p_2$ against $H_a : p_1 > p_2$ at level of significance at  = 0.01, and
compute the p-value.

(b) Use R commands to conduct the test of part (a) and to construct a 99% CI for
$p_1 - p_2$.

```python
import numpy as np
import scipy.stats as stats

# Given data
n1, x1 = 200, 16
n2, x2 = 400, 14

# Sample proportions
p1 = x1 / n1
p2 = x2 / n2

# Pooled proportion under H0
p_hat = (x1 + x2) / (n1 + n2)

# Standard error
SE = np.sqrt(p_hat * (1 - p_hat) * (1/n1 + 1/n2))

# Compute Z-test statistic
z_stat = (p1 - p2) / SE

# Compute one-tailed p-value for H_A: p1 > p2
p_value = 1 - stats.norm.cdf(z_stat)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Compute 99% Confidence Interval for p1 - p2
alpha = 0.01
z_crit = stats.norm.ppf(1 - alpha / 2)

CI_lower = (p1 - p2) - z_crit * np.sqrt(p1 * (1 - p1) / n1 + p2 * (1 - p2) / n2)
CI_upper = (p1 - p2) + z_crit * np.sqrt(p1 * (1 - p1) / n1 + p2 * (1 - p2) / n2)

print(f"99% Confidence Interval for (p1 - p2): ({CI_lower:.4f}, {CI_upper:.4f})")
```

```python
import pandas as pd
import scipy.stats as stats
import numpy as np

# Load the dataset
df = random_sample
#df = pd.read_csv("grades.csv")
# Separate GPA by gender
male_gpa = df[df["GENDER"] == "Male"]["GPA"]
female_gpa = df[df["GENDER"] == "Female"]["GPA"]

# Perform a two-sample t-test
t_stat, p_value = stats.ttest_ind(male_gpa, female_gpa, equal_var=False)  # Welchs t-test

# Compute 95% confidence interval for the difference in means
mean_diff = male_gpa.mean() - female_gpa.mean()
se_diff = np.sqrt(male_gpa.var(ddof=1) / len(male_gpa) + female_gpa.var(ddof=1) / len(female_gpa))
ci_lower = mean_diff - 1.96 * se_diff
ci_upper = mean_diff + 1.96 * se_diff

# Print results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"95% Confidence Interval for Mean GPA Difference: ({ci_lower:.4f}, {ci_upper:.4f})")

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in GPA between male and female students.")
else:
    print("There is no significant difference in GPA between male and female students.")
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# Load the dataset
df = pd.read_csv("grades.csv")

# Remove rows with missing values in relevant columns
df = df.dropna(subset=["GENDER", "MAT 151"])

# Count students who took MAT 151 (score > 0) for each gender
male_total = sum(df["GENDER"] == "Male")
female_total = sum(df["GENDER"] == "Female")

male_took_mat151 = sum((df["GENDER"] == "Male") & (df["MAT 151"] > 0))
female_took_mat151 = sum((df["GENDER"] == "Female") & (df["MAT 151"] > 0))

# Compute proportions
p_male = male_took_mat151 / male_total
p_female = female_took_mat151 / female_total

# Compute pooled proportion
p_pool = (male_took_mat151 + female_took_mat151) / (male_total + female_total)

# Compute standard error
se = np.sqrt(p_pool * (1 - p_pool) * (1/male_total + 1/female_total))

# Compute z-score
z_stat = (p_male - p_female) / se

# Compute p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Compute 95% confidence interval
ci_lower = (p_male - p_female) - 1.96 * se
ci_upper = (p_male - p_female) + 1.96 * se

# Print results
print(f"Proportion of males who took MAT 151: {p_male:.4f}")
print(f"Proportion of females who took MAT 151: {p_female:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"95% Confidence Interval for Difference in Proportions: ({ci_lower:.4f}, {ci_upper:.4f})")

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in the proportion of males and females who took MAT 151.")
else:
    print("There is no significant difference in the proportion of males and females who took MAT 151.")
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# Load the dataset
df = pd.read_csv("grades.csv")

# Remove rows with missing values in relevant columns
df = df.dropna(subset=["GENDER", "GPA"])

# Count students with GPA > 3.0 for each gender
male_total = sum(df["GENDER"] == "Male")
female_total = sum(df["GENDER"] == "Female")

male_above_3 = sum((df["GENDER"] == "Male") & (df["GPA"] > 3.0))
female_above_3 = sum((df["GENDER"] == "Female") & (df["GPA"] > 3.0))

# Compute proportions
p_male = male_above_3 / male_total
p_female = female_above_3 / female_total

# Compute pooled proportion
p_pool = (male_above_3 + female_above_3) / (male_total + female_total)

# Compute standard error
se = np.sqrt(p_pool * (1 - p_pool) * (1/male_total + 1/female_total))

# Compute z-score
z_stat = (p_male - p_female) / se

# Compute p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Compute 95% confidence interval
ci_lower = (p_male - p_female) - 1.96 * se
ci_upper = (p_male - p_female) + 1.96 * se

# Print results
print(f"Proportion of males with GPA > 3.0: {p_male:.4f}")
print(f"Proportion of females with GPA > 3.0: {p_female:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"95% Confidence Interval for Difference in Proportions: ({ci_lower:.4f}, {ci_upper:.4f})")

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in the proportion of males and females with GPA > 3.0.")
else:
    print("There is no significant difference in the proportion of males and females with GPA > 3.0.")
```

```python
import numpy as np
from scipy import stats

sample1 = np.array([78,85, 82, 76, 88, 90,84, 79])
sample2 = np.array([92, 89, 95, 87, 91, 94, 90, 88])
# Sample sizes
n1, n2 = len(sample1), len(sample2)

# Sample means and standard deviations
X1, X2 = np.mean(sample1), np.mean(sample2)
S1, S2 = np.std(sample1, ddof=1), np.std(sample2, ddof=1)

# Standard error
SE = np.sqrt(S1**2 / n1 + S2**2 / n2)

# Welch-Satterthwaite degrees of freedom
df_welch = ((S1**2 / n1 + S2**2 / n2) ** 2) / (((S1**2 / n1) ** 2) / (n1 - 1) + ((S2**2 / n2) ** 2) / (n2 - 1))

# Compute test statistic (t-value)
t_stat = (X1 - X2) / SE

# Compute one-tailed p-value for H_A: 1 > 2
p_value = 1 - stats.t.cdf(t_stat, df_welch)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Compute 90% Confidence Interval for 1 - 2
alpha = 0.05
t_crit = stats.t.ppf(1 - alpha / 2, df_welch)

CI_lower = (X1 - X2) - t_crit * SE
CI_upper = (X1 - X2) + t_crit * SE

print(f"90% Confidence Interval for mean difference: ({CI_lower:.4f}, {CI_upper:.4f})")
# expected 90% CI (1.052603, 1.652032)
```
