# Your answer: 
Multivariate Categorical Data: Multivariate categorical data involves more than one categorical variable in a dataset, where each observation or data point can be categorized into multiple groups simultaneously.

Population and Sample: The population refers to the entire set of individuals or instances that meet particular criteria. A sample is a subset of the population that is selected for analysis.

Confidence Interval: A confidence interval is a statistical tool used to estimate the range in which we are reasonably confident that a population parameter (e.g., mean or proportion) lies, based on a sample from that population.

Probability vs. Non-probability Sampling: Probability sampling involves random selection, where each member of the population has a known and equal chance of being included in the sample. Non-probability sampling does not rely on random selection and may lead to a biased sample.

Quantitative Variables and Qualitative Variables:

Quantitative Variables: These variables represent measurable quantities and can be expressed in numerical terms. Examples include age, height, weight, income, etc.
Qualitative Variables: Also known as categorical variables, these represent categories and cannot be measured in numerical terms. Examples include gender, color, nationality, etc.

#### Task 5: Calculate Confidence Interval for Single Population Mean. (0.1p)

You are conducting a study to determine the average time (in minutes) it takes for customers to complete a specific task on your website. You collect a random sample of 50 customer data points and find that the sample mean time is 8.5 minutes with a standard deviation of 1.2 minutes. Calculate a 95% confidence interval for the true average time it takes for customers to complete the task.

```python
import scipy.stats as stats

sample_mean = 8.5
std_dev = 1.2
sample_size = 50

confidence_level = 0.95

margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * (std_dev / (sample_size ** 0.5))

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Confidence Interval: {confidence_interval}")
```

#### Task 6: Calculate Confidence Interval for Single Population Proportion (0.1p)

You work for a survey company that conducts political polls. You've conducted a poll to estimate the proportion of registered voters who support a particular candidate in an upcoming election. From a random sample of 1,000 registered voters, you find that 540 express support for the candidate. Calculate a 90% confidence interval for the true proportion of registered voters who support the candidate.

```python
import statsmodels.api as sm

supporters = 540
total_voters = 1000

confidence_level = 0.9

proportion = supporters / total_voters

confidence_interval = sm.stats.proportion_confint(supporters, total_voters, alpha=1 - confidence_level)

print(f"Confidence Interval: {confidence_interval}")
```
